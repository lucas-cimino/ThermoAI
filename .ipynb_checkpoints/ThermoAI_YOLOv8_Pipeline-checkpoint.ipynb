```python
### Initialization
from ultralytics import YOLO
from pathlib import Path
from viz_utils import draw_boxes
import torch
import os


# ---- Environment Setup ----
PROJECT = Path("runs_thermoai")
PROJECT.mkdir(exist_ok=True)


DATA_YAML = "data_resolved.yaml"
RUN_NAME = "yolov8s_2cls"
MODEL_NAME = "yolov8s.pt"


TRAIN_DIR = Path("dataset/images/train")
VAL_DIR = Path("dataset/images/val")
TEST_DIR = Path("dataset/images/test")


LABEL_TRAIN_DIR = Path("dataset/labels/train")
LABEL_VAL_DIR = Path("dataset/labels/val")
LABEL_TEST_DIR = Path("dataset/labels/test")


# Device check
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"‚úÖ Using device: {device.upper()}")
```

```python
### Train
print("\nüöÄ Starting YOLOv8 Training Phase...")
model = YOLO(MODEL_NAME)


results = model.train(
data=DATA_YAML,
epochs=30,
imgsz=640,
batch=16,
project=str(PROJECT),
name=RUN_NAME,
device=device,
pretrained=True,
)


print("\nüèÅ Training completed. Metrics summary:")
print({
"mAP50-95": float(results.results_dict.get('metrics/mAP50-95', 0)),
"mAP50": float(results.results_dict.get('metrics/mAP50', 0)),
"precision": float(results.results_dict.get('metrics/precision', 0)),
"recall": float(results.results_dict.get('metrics/recall', 0)),
})
```

```python
### Validation
def find_best(project: Path, run_hint: str) -> Path:
    cand = project / run_hint / "weights" / "best.pt"
    if cand.exists():
        return cand
        runs = sorted([p for p in project.glob(f"{run_hint}*") if p.is_dir()], key=lambda p: p.stat().st_mtime, reverse=True)
    if runs:
        alt = runs[0] / "weights" / "best.pt"
        if alt.exists():
            return alt
    raise FileNotFoundError(f"No best.pt found under {project}/{run_hint}*")


print("\nüß™ Starting Validation Phase...")
best_pt = find_best(PROJECT, RUN_NAME)
model = YOLO(str(best_pt))


metrics_val = model.val(data=DATA_YAML, imgsz=640, split="val")
print("\nüìä Validation Metrics:")
print({
"mAP50-95": float(metrics_val.box.map),
"mAP50": float(metrics_val.box.map50),
"precision": float(metrics_val.box.mp),
"recall": float(metrics_val.box.mr),
})


# Visualization setup
output_dir_val = PROJECT / "eval_viz"
output_dir_val.mkdir(parents=True, exist_ok=True)


results_val = model.predict(source=str(VAL_DIR), imgsz=640, conf=0.25, save=False)
for r in results_val:
    img_path = Path(r.path)
    label_path = LABEL_VAL_DIR / (img_path.stem + ".txt")
    out_path = output_dir_val / img_path.name
    draw_boxes(str(img_path), str(label_path), r, str(out_path))


print(f"\nüñºÔ∏è Annotated validation images saved to: {output_dir_val}")
```

```python
### Test
print("\nüî¨ Starting Testing Phase...")
metrics_test = model.val(data=DATA_YAML, imgsz=640, split="test")
print("\nüìä Test Metrics:")
print({
"mAP50-95": float(metrics_test.box.map),
"mAP50": float(metrics_test.box.map50),
"precision": float(metrics_test.box.mp),
"recall": float(metrics_test.box.mr),
})


output_dir_test = PROJECT / "test_viz"
output_dir_test.mkdir(parents=True, exist_ok=True)


results_test = model.predict(source=str(TEST_DIR), imgsz=640, conf=0.25, save=False)
for r in results_test:
    img_path = Path(r.path)
    label_path = LABEL_TEST_DIR / (img_path.stem + ".txt")
    out_path = output_dir_test / img_path.name
    draw_boxes(str(img_path), str(label_path), r, str(out_path))


print(f"\nüñºÔ∏è Annotated test images saved to: {output_dir_test}")
print("\n‚úÖ Pipeline execution completed successfully.")
```
